# Canny edge detector

The Canny filter uses a Gaussian-based method to compute intensity gradients to eventually narrow down edges to a single pixel.
There are three adjustable parameters: the width of the Gaussian and the low and high threshold for the hysteresis thresholding step of the algorithm.

```python
import skimage as ski
import matplotlib.pyplot as plt
```

```python
cell = ski.data.cell()
cell.shape
```

```python
edges_s1 = ski.feature.canny(cell, sigma=1)
edges_s3 = ski.feature.canny(cell, sigma=3)
fig, axes = plt.subplots(ncols=3, figsize=(12, 4))
axes = axes.flatten()
for i, img in enumerate([cell, ~edges_s1, ~edges_s3]):
    axes[i].imshow(img)
    axes[i].set_axis_off()
plt.show()
```

```python
cell_eq = ski.exposure.adjust_sigmoid(cell)
edges = ski.feature.canny(cell_eq, sigma=3)
fig, axes = plt.subplots(ncols=3, figsize=(12, 4))
axes = axes.flatten()
for i, img in enumerate([cell, cell_eq, ~edges]):
    axes[i].imshow(img)
    axes[i].set_axis_off()
plt.show()
```
