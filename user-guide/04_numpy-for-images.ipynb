```python
import skimage as ski
import matplotlib.pyplot as plt
```

## Gray image

```python
camera = ski.data.camera()
type(camera), camera.shape
```

```python
ax = plt.imshow(camera)
ax.axes.set_axis_off()
plt.show()
```

```python
mask = camera < camera.mean()
camera_masked = camera.copy()
camera_masked[mask] = 0

fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(9, 3))
axes = axes.flatten()
axes[0].imshow(camera)
axes[1].imshow(mask)
axes[2].imshow(camera_masked)
for ax in axes:
    ax.axes.set_axis_off()

plt.show()
```
